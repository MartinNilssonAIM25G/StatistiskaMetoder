```python
import pandas as pd
import numpy as np

df = pd.read_csv("data/housing.csv")

df_model = df[["median_income", "housing_median_age", "total_rooms", "households", "ocean_proximity", "median_house_value"]].copy()
df_model["avg_rooms"] = (df_model["total_rooms"] / df_model["households"])
df_model = df_model.drop(columns=["total_rooms", "households"])

X = df_model.drop(columns=["median_house_value"])
y = df_model["median_house_value"]

X = pd.get_dummies(X, columns=["ocean_proximity"], drop_first=True)
X = X.astype(np.float64)  # <-- Konvertera allt till float64
print(X.dtypes)  # <-- Kolla datatyperna
print(X.head())
```

We select a subset of variables based on interpretability and expected relevance. Median income is known to be the strongest predictor of housing prices. Housing age and average room count reflect structural characteristics. Ocean proximity is included as a categorical variable to capture location effects.
